```python
import pandas as pd
import numpy as np
```

```python
lr=pd.read_csv('/content/drive/MyDrive/Python Datasets/Social_Network_Ads(1).csv')
```

```python
lr.head()
```

```python
lr.isnull().sum()
```

```python
# Split data into dependent and independent variables
X=lr.iloc[:,[1,2,3]].values
y=lr.iloc[:,4].values
```

```python
print(X)
```

```python
# Convert Categorical variables into numerical Variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,0]=le.fit_transform(X[:,0])
```

```python
# Split data into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
print(X_train)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#implement logistic regression
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
#test the model- create confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# Measure Accuracy, Precision, Recall and F1 Score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

# Gaussian NB algorithm

```python
# implement gaussian NB algo
from sklearn.naive_bayes import GaussianNB
classifierNB=GaussianNB()
classifierNB.fit(X_train,y_train)
```

```python
y_predNB=classifierNB.predict(X_test)
```

```python
# Confusion Matrix for Gaussion NB
from sklearn.metrics import confusion_matrix
cmNB=confusion_matrix(y_test,y_predNB)
print(cmNB)
```

```python
# Measure Accuracy, Precision, recall and f1 score for Gaussion NB
accuracyNB = accuracy_score(y_test, y_predNB)
precisionNB = precision_score(y_test, y_predNB)
recallNB = recall_score(y_test, y_predNB)
f1NB = f1_score(y_test, y_predNB)
print("AccuracyNB:", accuracyNB)
print("PrecisionNB:", precisionNB)
print("RecallNB:", recallNB)
print("F1 ScoreNB:", f1NB)
```

```python
# Compare results of Logistic regression and gaussion NB algo
print("Logistic Regression Results:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nGaussian NB Results:")
print("AccuracyNB:", accuracyNB)
print("PrecisionNB:", precisionNB)
print("RecallNB:", recallNB)
print("F1 ScoreNB:", f1NB)
```
