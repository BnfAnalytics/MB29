```python
# Split data into dependent and independent variables
X=df1.iloc[:,[3,4]].values
print(X.shape)
```

```python
# Model building using random number for clustering
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,init='k-means++',max_iter=300,n_init=10,random_state=41)
```

```python
km.fit(X)
```

```python
km.labels_
```

```python
y_means=km.fit_predict(X)
```

```python
X[y_means==0,0]
```

```python
# Create a plot of K Mean Clusters using scatter plot
plt.figure(figsize=(10,5))
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('K Mean Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Age')
plt.legend()
plt.show()
```

```python
# Silhouette_Score compute the mean Silhouette Coefficient of all samples
from sklearn.metrics import silhouette_score
silhouette_avg=silhouette_score(X,y_means)
print(silhouette_avg)
# as high as the value more perfect the clusters are created
```

```python
# Taking the range for centroids from 1 to 10-elbow method
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,
                init='k-means++',
                max_iter=300,
                n_init=10,
                random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
# Plot Elbow results
plt.figure(figsize=(10,5))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# As per Elbow method we are getting 5 as most optimized number of clusters, so implement it on our created model
km=KMeans(n_clusters=5,
          init='k-means++',
          max_iter=300,
          n_init=10,
          random_state=0)
y_means=km.fit_predict(X)
```

```python
silhouette_avg=silhouette_score(X,y_means)
print(silhouette_avg)
```

```python
# plot scatter for this clusters
plt.figure(figsize=(10,5))
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(X[y_means==3,0],X[y_means==3,1],s=100,c='cyan',label='Cluster 4')
plt.scatter(X[y_means==4,0],X[y_means==4,1],s=100,c='magenta',label='Cluster 5')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('K Mean Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Age')
plt.legend()
plt.show()
```
